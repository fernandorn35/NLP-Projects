```python
skipped_accounts = []
success_accounts = []
tweet_database = {}
counter = 0
for name in tqdm(names[0:10]):
    counter += 1
    if counter >= 1000:
        counter = 0
        renew_ip()
        
    try:
        tweets = get_weibo_tweets_by_name(name=name, pages=1)
        for tweet in tweets:
            tweet_database[name] = tweet
        success_accounts.append(name)
    except:
        skipped_accounts.append(name)
```
